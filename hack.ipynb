```python
df1 = pd.read_csv("articles.csv", encoding='utf-8')
df1 = df1.dropna()
df1['id_model'] = range(len(df1))
df2 = pd.read_csv("life_situations.csv", encoding='utf-8')
df2 = df2.dropna()
df2['id_model'] = range(95667, 95667 + len(df2))
df3 = pd.read_csv("news.csv", encoding='utf-8')
df3.pop("short_description")
df3 = df3.dropna()
df3['id_model'] = range(100066, 100066 + len(df3))
df4 = pd.read_csv("services.csv", encoding='utf-8')
df4 = df4.dropna()
df4['id_model'] = range(195653, 195653 + len(df4))
```

```python
df1['text'] = df1['title'] + " [SEP] " + df1['content']+" [SEP] "+df1['sys_lang']
df2['text'] = df2['title_main'] + " [SEP] "+df2['title_sub'] +" [SEP] "+df2['intro']+" [SEP] "+ df2['instruction']+" [SEP] "+df2['sys_lang']
df3['text'] = df3['title'] + " [SEP] " + df3['body']+ " [SEP] "+df3['sys_lang']
df4['text'] = df4['description'] +" [SEP] " + df4['full_title']+ " [SEP] " + df4['short_title']+" [SEP] " + df4['additional_info']+" [SEP] "+  df4['result_description']+" [SEP] "+df4['sys_lang']
data_texts1 = df1['text'].to_list()
data_texts2 = df2['text'].to_list()
data_texts3 = df3['text'].to_list()
data_texts4 = df4['text'].to_list()
```

```python
df2
```

```python
concatenated_data = []
for data_texts in [data_texts1, data_texts2, data_texts3, data_texts4]:
    concatenated_data.extend(data_texts)
```

```python
import torch
dev = torch.device("cuda")
```

```python
model = SentenceTransformer('LaBSE').to(dev)
```

```python
corpus_embeddings = model.encode(concatenated_data, show_progress_bar=True, convert_to_tensor=True)
```

```python
torch.save(model.state_dict(), "db.pth")
```

```python
from langdetect import detect
def detect_language(text):
    try:
        detected_language = detect(text)
        return detected_language
    except Exception as e:
        return "Unknown"  # If language detection fails
```

```python
predictions = []
id_q=0
for question in test_questions:
    id_q+=1
    detector=detect_language(question)
    question_embedding = model.encode(question, convert_to_tensor=True)
    hits = util.semantic_search(question_embedding, corpus_embeddings, top_k=1)  # To get top 1 hit
    for i in range(len(hits)):
        if len(df1[df1['id_model']==hits[0][0]['corpus_id']])==1:
            if (df1[df1['id_model']==hits[0][0]['corpus_id']].values[i][2])==detector:
                id=df1[df1['id_model']==hits[0][0]['corpus_id']].values[i][0]
                break
        if len(df2[df2['id_model']==hits[0][0]['corpus_id']])==1:
            if (df2[df2['id_model']==hits[0][0]['corpus_id']].values[i][2])==detector:
                id=df2[df2['id_model']==hits[0][0]['corpus_id']].values[i][0]
                break
        if len(df3[df3['id_model']==hits[0][0]['corpus_id']])==1:
            if (df3[df3['id_model']==hits[0][0]['corpus_id']].values[i][2])==detector:
                id=df3[df3['id_model']==hits[0][0]['corpus_id']].values[i][0]
                break
        if len(df4[df4['id_model']==hits[0][0]['corpus_id']])==1:
            if (df4[df4['id_model']==hits[0][0]['corpus_id']].values[i][2])==detector:
                id=df4[df4['id_model']==hits[0][0]['corpus_id']].values[i][0]
                break
    predictions.append((id_q, id))    
```

```python
submission_df = pd.DataFrame(predictions)
```
